```python
df1 = df.drop_duplicates(subset=['course_name','instructor']).reset_index(drop=True)
df1.head()
```

The selected features for computing similarities are 
1. course name
2. instructor
3. certification offered
4. study material available
5. difficulty level
6. course price

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from scipy.sparse import hstack
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Text Features
tfidf = TfidfVectorizer(stop_words='english', ngram_range=(1,2), max_features=5000)
text_features = tfidf.fit_transform(df1['course_name'] + " " + df1['instructor'])
```

```python
# Encoding categorical features
cat_features = pd.get_dummies(df1[['certification_offered', 'study_material_available']])
ordinal = OrdinalEncoder()
difficulty = ordinal.fit_transform(df1[['difficulty_level']])
```

```python
# Standardize numerical values
scaler = StandardScaler()
num_features = scaler.fit_transform(df1[['course_price']])
```

```python
# Combine all
matrix = hstack([text_features, cat_features.values, difficulty, num_features])
```

```python
# Computing cosine similarity
cosine = cosine_similarity(matrix)
```

```python
def recommend_course(course_name, top_n=5):
# Check whether the course exists
    if course_name not in df1['course_name'].values:
        print("Course not found!")
        return
# Find the index of the selected course
    idx = df1[df1['course_name'] == course_name].index[0]
# Get similarity scores
    sim_scores = list(enumerate(cosine[idx]))
# Sort based on similarity
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    recommendations = []
# Skip the first one (same course)
    for i, score in sim_scores:
        course = df1.iloc[i]['course_name']
        if course == course_name:
            continue
        if course not in recommendations:
            recommendations.append(course)
        if len(recommendations) == top_n:
            break
# Avoid duplicate course names
        if course not in recommendations:
            recommendations.append(course)
# Stop after top_n recommendations
        if len(recommendations) == top_n:
            break
    print(f"\nRecommended courses similar to '{course_name}'")
    print("-" * 50)
    for n, course in enumerate(recommendations, start=1):
        print(f"{n}. {course}")
```

# KNN Recommendation System Model

```python
from sklearn.neighbors import NearestNeighbors
from scipy.sparse import hstack
```

```python
course_features = hstack([
    text_features,
    num_features,
    cat_features
]).tocsr()
#Here we combine TF-IDF Features, Encoded Categories,Numerical Features into one feature matrix.
```

```python
#Train KNN
knn_model = NearestNeighbors(
    n_neighbors=6,
    metric='cosine',
    algorithm='brute'
)

knn_model.fit(course_features)
```

```python
#Recommendation Function
def recommend_knn(course_name, n_recommendations=5):

    # Check if course exists
    if course_name not in df1['course_name'].values:
        return "Course not found"

    # Find index
    idx = df1[df1['course_name'] == course_name].index[0]

    # Find neighbours
    distances, indices = knn_model.kneighbors(
        course_features[idx:idx+1],
        n_neighbors=n_recommendations + 1
    )

    recommendations = []

    for i in indices.flatten()[1:]:
        recommendations.append({
            "Course": df1.iloc[i]['course_name'],
            "Instructor": df1.iloc[i]['instructor'],
            "Certificate": df1.iloc[i]['certification_offered'],
            "Rating": df1.iloc[i]['course_price']
        })

    return pd.DataFrame(recommendations)
```

```python
# Number of neighbours (excluding itself)
k = 5
# Get distances and indices
distances, indices = knn_model.kneighbors(
    course_features,
    n_neighbors=k + 1
)
# Remove self-neighbour (distance = 0)
distances = distances[:, 1:]
# Convert cosine distance to cosine similarity
similarities = 1 - distances
# Calculate average similarity
average_similarity = np.mean(similarities)

print(f"Average Similarity Score: {average_similarity:.4f}")
```

# Collaborative Filtering Model

```python
np.random.seed(42)
n_users = 500
interactions = []

for user in range(1, n_users + 1):
    watched_courses = np.random.choice(
        df1.index,
        size=np.random.randint(5,15),
        replace=False
    )

    for course in watched_courses:
        interactions.append([
            user,
            course,
            np.random.randint(3,6)
        ])

interaction_df = pd.DataFrame(
    interactions,
    columns=['user_id','course_id','rating']
)
```

```python
#Create User Item Matrix
user_item_matrix = interaction_df.pivot_table(
    index='user_id',
    columns='course_id',
    values='rating'
).fillna(0)
```

```python
#Train KNN Collaborative Filtering
cf_model = NearestNeighbors(
    metric='cosine',
    algorithm='brute'
)
cf_model.fit(user_item_matrix.T)
```

```python
#Recommendation Function
def collaborative_recommend(course_name, n_recommendations=5):

    if course_name not in df1['course_name'].values:
        return "Course not found"

    course_index = df1[df1['course_name']==course_name].index[0]
    distances, indices = cf_model.kneighbors(
        user_item_matrix.T.iloc[course_index].values.reshape(1,-1),
        n_neighbors=n_recommendations+1
    )

    recommendations = []

    for idx in indices.flatten()[1:]:
        recommendations.append({
            "Course": df1.iloc[i]['course_name'],
            "Instructor": df1.iloc[i]['instructor'],
            "Certificate": df1.iloc[i]['certification_offered'],
            "Rating": df1.iloc[i]['course_price']
        })

    return pd.DataFrame(recommendations)
```
